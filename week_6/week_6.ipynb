## NumPy Arrays: Storing Multiple Values

So far we've stored single values in variables. But for our expense tracker, we need to store **many expenses** - not just one coffee, but a whole week's worth of spending! This is where **collections** come in.

While Python has built-in **lists** (created with `[]`), for numerical data we prefer **NumPy arrays**. NumPy is the foundation of scientific computing in Python.

NumPy is an **external library** — it doesn't come with Python itself, so we need to **import** it before using it. Google Colab has it pre-installed.

> ℹ️ **Reminder:** If you're working locally, you may need to install external libraries first by running `pip install <library_name>` in your terminal (e.g., `pip install numpy`).

**Why NumPy over lists?**
- Much faster for numerical operations
- Supports element-wise operations (no loops needed!)
- Foundation for Pandas, Matplotlib, and most data science tools
- Standard in neuroscience (MNE-Python, NiLearn, etc.)

````python
# Import NumPy - conventionally abbreviated as 'np'
import numpy as np

# Create a NumPy array of expense amounts
amounts = np.array([12.50, 8.00, 15.75, 25.00, 9.25])
print("Expense amounts:", amounts)
print("Type:", type(amounts))

# For text data, we still use regular Python lists
categories = ["Food", "Transport", "Food", "Entertainment", "Food"]
print("\nCategories:", categories)
````

````python
# Basic operations work like lists
print("Number of expenses:", len(amounts))
print("First expense:", amounts[0])
print("Last expense:", amounts[-1])

# But NumPy gives us powerful built-in functions!
print("\n--- NumPy superpowers ---")
print("Sum:", amounts.sum())           # No loop needed!
print("Mean:", amounts.mean())
print("Max:", amounts.max())
print("Min:", amounts.min())

# Element-wise operations - apply to ALL elements at once
print("\nDouble all amounts:", amounts * 2)
print("Add £5 to each:", amounts + 5)
````

````python
# Adding items to a NumPy array (and keeping categories in sync)
amounts = np.append(amounts, 12.00)
categories.append("Transport")
print("After adding £12.00 (Transport):", amounts)

amounts = np.append(amounts, 30.00)
categories.append("Entertainment")
print("After adding £30.00 (Entertainment):", amounts)

print(f"\nNow we have {len(amounts)} expenses")
print(f"Categories: {categories}")
print(f"New total: £{amounts.sum():.2f}")
````

# Part 2: Control Flow and Dictionaries

Now that we can store data, we need to **process** it. Control flow lets us:
- **Loop** through data (do something repeatedly)
- **Make decisions** (do different things based on conditions)

## For Loops

A `for` loop lets us go through each item in a collection, one at a time.

### Example: Reviewing All Your Expenses

**The scenario:** It's the end of the week and you want to review your spending. You have 7 expenses stored in your arrays, but staring at raw data like `[12.50, 8.00, 15.75, ...]` isn't very helpful.

**What we want:** Print each expense nicely, showing both the amount AND what category it belongs to.

**The solution:** Use a `for` loop with an index variable `i` to access both lists at the same position:
- When `i = 0`: get `amounts[0]` and `categories[0]` (the first expense)
- When `i = 1`: get `amounts[1]` and `categories[1]` (the second expense)
- ...and so on

````python
# Let's reset our expense data as NumPy array
amounts = np.array([12.50, 8.00, 15.75, 25.00, 9.25, 12.00, 30.00])
categories = ["Food", "Transport", "Food", "Entertainment", "Food", "Transport", "Entertainment"]

# Loop through using index numbers (0, 1, 2, ...)
# range(len(amounts)) gives us: 0, 1, 2, 3, 4, 5, 6
for i in range(len(amounts)):
    print(f"{categories[i]}: £{amounts[i]}")
````

### Example: Tracking Your Running Total

**The scenario:** You want to see **how your spending adds up over the week**. After each expense, how much have you spent so far?

**What we want:** Calculate a running total that grows with each expense — like checking your bank balance after each purchase.

````python
# Calculate total spending step by step
total = 0  # Start with nothing

for amount in amounts:
    total = total + amount  # Add this expense to the running total
    print(f"Added £{amount:.2f} → Running total: £{total:.2f}")

print(f"\nFinal total: £{total:.2f}")
````

### Example: Repeating an Action N Times

**The scenario:** Sometimes you don't have a collection to loop through — you just need to do something a **specific number of times**.

**What we want:** Simulate saving £100 every month for 6 months, with compound interest. We need to repeat the same calculation 6 times.

````python
# Simulate saving with compound interest
# You deposit £100 each month, and earn 2% interest on your balance

monthly_saving = 100
interest_rate = 0.02  # 2% per month
balance = 0
num_months = 6

for month in range(1, num_months + 1):
    # First, add this month's deposit
    balance = balance + monthly_saving
    # Then, apply interest on the total balance
    interest_earned = balance * interest_rate
    balance = balance + interest_earned
    print(f"Month {month}: Deposit £{monthly_saving}, Interest £{interest_earned:.2f} → Balance: £{balance:.2f}")

print(f"\nAfter {num_months} months: £{balance:.2f}")
print(f"Total deposited: £{monthly_saving * num_months}")
print(f"Total interest earned: £{balance - (monthly_saving * num_months):.2f}")
````

## If Statements: Making Decisions

Sometimes we need our code to **do different things** depending on the data. `if` statements let us make decisions based on conditions.

````python
# Simple if statement
expense = 25.00

if expense > 20:
    print("This is a large expense!")
else:
    print("This is a reasonable expense.")
````

````python
# Using if/elif/else for multiple conditions
expense = 15.00

if expense < 10:
    size = "small"
elif expense < 20:
    size = "medium"
else:
    size = "large"

print(f"£{expense} is a {size} expense")
````

````python
# Combining loops and if statements to classify all expenses
for amount in amounts:
    if amount < 10:
        size = "small"
    elif amount < 20:
        size = "medium"
    else:
        size = "large"
    
    print(f"£{amount:.2f} - {size}")
````

## Dictionaries: Key-Value Storage

A **dictionary** stores data as key-value pairs. Think of it like a real dictionary: you look up a word (key) to find its definition (value).

Dictionaries use curly braces `{}` with colons between keys and values:

````python
# Creating a dictionary
expense_record = {
    "date": "2026-01-10",
    "category": "Food",
    "amount": 12.50,
    "description": "Lunch"
}

print(expense_record)
````

````python
# Accessing values using keys
print("Category:", expense_record["category"])
print("Amount:", expense_record["amount"])

# Adding a new key-value pair
expense_record["paid_with"] = "card"
print(expense_record)
````

### Example: Summing Expenses by Category

**The scenario:** We have 7 expenses spread across 3 categories (Food, Transport, Entertainment). We want to know how much we spent on each category.

**What we want:** Add up all the Food expenses together, all the Transport expenses together, and so on.

**Why a dictionary?** We can use the category name as the key and keep a running total as the value:
- `{"Food": 37.50, "Transport": 20.00, "Entertainment": 55.00}`

````python
# Reminder of our data
print("Amounts:", amounts)
print("Categories:", categories)
````

````python
# Summing expenses by category
totals = {}  # Start with an empty dictionary

# We need to loop through both lists at the same time
# zip() pairs up items from multiple lists
for category, amount in zip(categories, amounts):
    if category in totals:
        # Category already exists, add to it
        totals[category] += amount
    else:
        # New category, create it
        totals[category] = amount

print("Totals by category:", totals)
````

### Looping Through a Dictionary

**What we want:** Display our totals nicely. How do we loop through a dictionary?

There are three ways:
- **Keys only:** `for category in totals:`
- **Values only:** `for amount in totals.values():`
- **Both keys and values:** `for category, amount in totals.items():`

The `.items()` method gives us pairs of (key, value) — perfect for printing a summary:

````python
# Iterating over a dictionary with .items()
print("Expense Summary:")
print("-" * 25)

for category, total in totals.items():
    print(f"Total spent on {category}: £{total:.2f}")
````

# Part 3: Functions

## Why Functions?

Notice how we keep writing similar code? For example, we've classified expenses as "small/medium/large" multiple times. What if we could write that logic **once** and reuse it everywhere?

A **function** is a named block of code that you can call whenever you need it. Functions help you:
- **Avoid repeating yourself** — write once, use many times
- **Organise your code** — break complex tasks into logical pieces
- **Make testing easier** — test each piece independently

Functions are defined with the `def` keyword:

````python
# A function that returns a value
def classify_amount(amount):
    """Classify an expense as small, medium, or large."""
    if amount < 10:
        return "small"
    elif amount < 20:
        return "medium"
    else:
        return "large"

# Use the function
result = classify_amount(15.00)
print(f"£15.00 is a {result} expense")

result = classify_amount(5.00)
print(f"£5.00 is a {result} expense")
````

### Default Arguments

You can give parameters default values, making them optional:

````python
# Function with default arguments
def classify_amount(amount, small_threshold=10, large_threshold=20):
    """Classify an expense with customisable thresholds."""
    if amount < small_threshold:
        return "small"
    elif amount < large_threshold:
        return "medium"
    else:
        return "large"

# Using default thresholds
print(classify_amount(15))

# Using custom thresholds
print(classify_amount(15, small_threshold=5, large_threshold=15))
````

````python
for amount in amounts:
    print(f"Current amount: {amount} -> {classify_amount(amount)}")
````

### Example: Creating a Summary Report

**The scenario:** Your friend asks: *"Can you give me a file showing how much I spent on each category this year?"*

**What we want:** 
1. Calculate total spending per category
2. Save the results to a CSV file they can open in Excel

**The solution:** A simple Python workflow:
1. **Load** the raw expense data (which we just did)
2. **Process** it — calculate the total for each category
3. **Save** the summary to a new file

This is exactly what researchers do: load raw experimental data, process it, and save the cleaned results for further analysis or sharing.

````python
# Calculate category totals from our CSV data
# Remember: all_rows[0] is the header, data starts at all_rows[1]
# Each row is: [Date, Day_of_Week, Category, Amount]

totals = {}
for row in all_rows[1:]:  # Skip the header row
    category = row[2]      # Category is column 3 (index 2)
    amount = float(row[3]) # Amount is column 4 (index 3), convert to number
    
    if category in totals:
        totals[category] += amount
    else:
        totals[category] = amount

print("Category totals:")
for category, total in totals.items():
    print(f"  {category}: £{total:.2f}")
````

Now let's **save** these totals to a new CSV file using `csv.writer`:

````python
# Save the category totals to a new CSV file
output_file = os.path.join(BASE_PATH, "sample_data", "category_totals.csv")

with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the header row
    writer.writerow(["Category", "Total"])
    
    # Write each category and its total
    for category, total in totals.items():
        writer.writerow([category, f"{total:.2f}"])

print(f"✓ Saved totals to: {output_file}")
````

## Classes: Bundling Data and Behaviour

**The problem:** So far, we've been storing expense data in different ways — lists, dictionaries, CSV rows. Each time we want to do something with an expense (check if it's large, format it nicely, categorise it), we write the same logic again.

**The scenario:** Your friend liked your expense summary so much that they want you to build them a proper expense tracking app. They want to:
- See if an expense is "essential" (like groceries, housing) or "discretionary" (like entertainment)
- Get warnings when expenses exceed their budget threshold
- Format expenses nicely for reports

You *could* write separate functions for all of this... but the logic would be scattered everywhere. What if there was a way to bundle the expense data *with* all the operations you can do on it?

**The solution: Classes**

A **class** is a blueprint for creating objects that bundle:
- **Data** (attributes) — the expense's date, category, amount
- **Behaviour** (methods) — functions that know how to work with that data

Think of it like this: a list of expenses is just raw data. An `Expense` object *understands* itself — it knows how to describe itself, categorise itself, and check if it's over budget.

````python
# Defining an Expense class
class Expense:
    # We createa a class-level variable: categories tagging the one that are considered "essential"
    ESSENTIAL_CATEGORIES = ["Housing", "Groceries", "Utilities", "Health/Medical", "Transportation"]
    
    # This is commonly called the "constructor" of a class.
    # This function must be always present and when not all of the variables that are defining 
    # inside the class will be initialised to their default value
    def __init__(self, date, day_of_week, category, amount):
        """Initialise an Expense with its details."""
        self.date = date
        self.day_of_week = day_of_week
        self.category = category
        self.amount = amount
    
    def is_essential(self):
        """Check if this expense is for an essential category."""
        return self.category in self.ESSENTIAL_CATEGORIES
    
    def is_over_budget(self, threshold=50):
        """Check if this expense exceeds a budget threshold."""
        return self.amount > threshold
    
    def get_warning(self, threshold=50):
        """Return a warning message if the expense is over budget."""
        if self.is_over_budget(threshold):
            return f"Warning: {self.category} expense of £{self.amount:.2f} exceeds £{threshold} budget!"
        return None
    
    def describe(self):
        """Return a formatted description of this expense."""
        essential_tag = "[Essential]" if self.is_essential() else "[Discretionary]"
        return f"{self.date} ({self.day_of_week}): £{self.amount:.2f} - {self.category} {essential_tag}"
    
    def to_dict(self):
        """Convert expense to a dictionary"""
        return {
            "date": self.date,
            "day_of_week": self.day_of_week,
            "category": self.category,
            "amount": self.amount,
            "is_essential": self.is_essential()
        }
````

````python
# Creating Expense objects
groceries = Expense("2025-01-08", "Wednesday", "Groceries", 195.79)
concert = Expense("2025-01-05", "Sunday", "Entertainment", 177.65)
coffee = Expense("2025-01-03", "Friday", "Dining Out", 12.50)

# The describe() method knows how to format itself AND classify itself
print("=== Expense Descriptions ===")
print(groceries.describe())
print(concert.describe())
print(coffee.describe())

# Check if expenses are essential or discretionary
print("\n=== Essential vs Discretionary ===")
print(f"Groceries essential? {groceries.is_essential()}")
print(f"Concert essential? {concert.is_essential()}")

# Budget warnings - the object knows how to check itself
print("\n=== Budget Warnings (threshold: £100) ===")
for expense in [groceries, concert, coffee]:
    warning = expense.get_warning(threshold=100)
    if warning:
        print(warning)
    else:
        print(f"✓ {expense.category}: £{expense.amount:.2f} is within budget")

# Convert to dictionary (useful for saving)
print("\n=== As Dictionary ===")
print(groceries.to_dict())
````

### Example: Cumulative Spending Over Time

**The scenario:** You want to visualise how your spending accumulates month by month throughout the year.

**What we want:** A line chart showing cumulative spending — so you can see how quickly you're burning through your annual budget.

````python
# Calculate monthly totals from our CSV data
# Remember: each row is [Date, Day_of_Week, Category, Amount]
# Date format is "2025-01-15" so we can extract year-month from the first 7 characters

monthly_totals = {}

for row in all_rows[1:]:  # Skip header
    date = row[0]
    month = date[5:7]  # Extract "2025-01" from "2025-01-15"
    amount = float(row[-1])
    
    if month in monthly_totals:
        monthly_totals[month] += amount
    else:
        monthly_totals[month] = amount

# Get months in order and their totals
months = list(monthly_totals.keys())
totals = list(monthly_totals.values())

print("Monthly spending:")
for month, total in monthly_totals.items():
    print(f"  {month}: £{total:.2f}")

# Calculate cumulative sum across months
cumulative = []
running_total = 0

for total in totals:
    running_total += total
    cumulative.append(running_total)

print(f"\nCumulative totals by month:")
for i, month in enumerate(months):
    print(f"  {month}: £{cumulative[i]:.2f}")

print(f"\nYear total: £{cumulative[-1]:.2f}")
````

````python
# Create the cumulative spending plot by month
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.figure(figsize=(10, 6))
plt.plot(month_names, cumulative, marker='o', linewidth=2, markersize=8, color='orange')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Cumulative Spending (£)", fontsize=12)
plt.title("How Spending Grows Over the Year", fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
````

### Example: Spending by Category

**The scenario:** You want to compare how much you spent across different categories.

**What we want:** A bar chart where each bar represents a category, making it easy to spot which categories dominate your budget.

````python
# First, calculate totals by category from our CSV data
category_totals = {}
for row in all_rows[1:]:  # Skip header
    category = row[2]      # Category is column 3
    amount = float(row[-1]) # Amount is column 4
    
    if category in category_totals:
        category_totals[category] += amount
    else:
        category_totals[category] = amount

# Bar chart of category totals
categories = list(category_totals.keys())
amounts = list(category_totals.values())

plt.figure(figsize=(12, 6))  # Wider to fit all category names
plt.bar(categories, amounts)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Spending (£)", fontsize=12)
plt.title("Spending by Category", fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate labels so they don't overlap

# Add value labels on top of each bar
for i, amt in enumerate(amounts):
    plt.text(i, amt + 100, f"£{amt:.0f}", ha='center', fontsize=9)

plt.tight_layout()  # Prevent labels from being cut off
plt.show()
````
